### Creando un archivo CSV mediando el uso del archivo JSON
2. Ignoramos el archivo `yelp.csv` y construye un DataFrame de `yelp.json`. Esto implica leer los datos en Python, decodificar el JSON, convertirlo en un DataFrame y agregar columnas individuales para cada uno de los tipos de votos.

````python
import csv, json, sys
inputFile = open("./yelp.json") #abrir el archivo json
outputFile = open("./cargar.csv", 'w') #cargar en el archivo csv

data = []
with open('./yelp.json') as f:
    for line in f:
        data.append(json.loads(line))
inputFile.close() #cerrar el archivo de entrada

output = csv.writer(outputFile) #crea a csv.write

output.writerow(data[0].keys())  # fila de cabecera

for row in data:
    output.writerow(row.values()) # valores de las filas
````

````python
print("Mostrando los datos aleatorios con el método \"Sample\"")
df_dic = pd.DataFrame(data)
df_dic.sample(6)
````

### Calculando el RMSE del Modelo a usar

6. Evalua el modelo dividiendo los datos en  conjuntos de entrenamiento y prueba y calculando el RMSE. ¿El RMSE tiene un sentido intuitivo para ti?



````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2 , random_state=70)
print("RMSE: %.2f" % np.mean((Reg_Lineal.predict(X_test) - y_test) ** 2))
````

Como sabemos el RMSE es el Error cuadrático medio, por lo tanto si este valor se aproxima a la recta horizontal de la variable Irreductible (Var(e)) se dice que es mas probable de predecir los target y pues vemos que el error relativo es 0.37, este valor es la suma entre varianza y el sesgo.

7. Intenta eliminar algunas de las características y analizar  si el RMSE mejora.
   

````python
def RMSE(X,Y):
    Reg_Lineal = LinearRegression()
    Reg_Lineal.fit(X,Y)
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2 , random_state=70)
    return np.mean((Reg_Lineal.predict(X_test) - y_test) ** 2)
    
U_C_df_Analizar= df_Analizar[["useful","cool"]]
F_C_df_Analizar= df_Analizar[["funny","cool"]]
F_U_df_Analizar= df_Analizar[["funny","useful"]]
print("Eliminando una característica\n")
print("Funny : \t{0:.2f}".format(RMSE(U_C_df_Analizar,Y)))
print("Useful : \t{0:.2f}".format(RMSE(F_C_df_Analizar,Y)))
print("Cool : \t\t{0:.2f}".format(RMSE(F_U_df_Analizar,Y)))

print("\nAnalizando con una característica\n")
F_df_Analizar= df_Analizar[["funny"]]
U_df_Analizar= df_Analizar[["useful"]]
C_df_Analizar= df_Analizar[["cool"]]

print("Funny : \t{0:.2f}".format(RMSE(F_df_Analizar,Y)))
print("Useful : \t{0:.2f}".format(RMSE(U_df_Analizar,Y)))
print("Cool : \t\t{0:.2f}".format(RMSE(C_df_Analizar,Y)))
````

10. En lugar de tratar este problema por regresión, como sería si fuese un problema de clasificación y qué precisión en el conjunto prueba se puede lograr con KNN.
   

````python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=60)

train = []
test = []

for i in range(1,10,1):
    KNC= KNeighborsClassifier(n_neighbors=i)    
    KNC.fit(X_train,y_train)
    train.append(KNC.score(X_train,y_train))
    test.append(KNC.score(X_test,y_test))
    
plt.plot(range(1,10),train)
plt.plot(range(1,10),test)
plt.show()
````

Cuando K esta entre [2 - 8] este modelo es impreciso, sin embargo en los demás valores es cercano a la recta de entrenamiento, es decir es mas predecible
